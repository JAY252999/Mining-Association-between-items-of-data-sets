# Apriori Algorithm

```python
#considering min 10% of rules by min_support function for analysing rules
frequent_itemsets = apriori(Book, min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
#Using Lift ratio metric for association rules
rules = association_rules(frequent_itemsets,metric="lift", min_threshold= 1)
rules
```

```python
# A high conviction value means that the consequent is highly depending on the antesecedent and range[0 inf]
```

```python
#Sorting rules
rules.sort_values('lift',ascending=False)[0:20]
```
