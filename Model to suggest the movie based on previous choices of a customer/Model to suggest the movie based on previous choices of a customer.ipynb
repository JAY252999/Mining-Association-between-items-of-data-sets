```python
#Importing all basic libraries and Assiciation functions from libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np
import scipy
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#reading data
Movies = pd.read_csv('my_movies.csv')
Movies
```

# Pre-processing

```python
#Creating dummy data for extracting association
Data_Dummies = pd.get_dummies(Movies)
Data_Dummies
```

# Apriori Algorithm

```python
#considering min 10% of rules by min_support function for analysing rules
frequent_itemsets = apriori(Data_Dummies,min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
#Using Lift ratio metric for association rules
Assiciation_rules = association_rules(frequent_itemsets, metric = "lift", min_threshold=0.7)
Assiciation_rules
```

```python
# A high conviction value means that the consequent is highly depending on the antecedent and range[0-inf]
```

```python
#Sorting rules
Assiciation_rules.sort_values('lift',ascending = False)[0:20]
```
